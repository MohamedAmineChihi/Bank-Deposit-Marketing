### Libraries

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

## Reading and inspecting the data 

```python
path = "../data/raw/socio_eco.csv"
se_df = pd.read_csv(path,  encoding='ANSI',sep=';',index_col=0)
#display(se_df) 
```

```python
#se_df.info()
```

### First impressions:
- We have **30** entries (rows). Each row represents a snapshot of different economical indicators at the end of each month between May 2008 and October 2010.
- We have a total of **4** features (columns). \
They represent economical indicators of each month:
* `tx_var_emploi` :
* `idx_prix_conso` :
* `idx_conf_conso` : 



### Next steps :

Even before exploring the data more thoroughly, it is clear that we will need to take care of a few things during pre-processing to clean the data and make it useable for our in-depth Analysis :
- We have to take care of the "???" values and convert date column to datetime.

# Data preprocessing 

```python
# Convert the 'date' column to datetime
se_df['date'] = pd.to_datetime(se_df['date'], dayfirst=True)

# Handling the missing values
# Replace "???" with NaN
# Replace None with NaN

se_df = se_df.replace('???', float('nan'))
# Convert columns to numeric
se_df['tx_var_emploi'] = pd.to_numeric(se_df['tx_var_emploi'])
se_df['idx_prix_conso'] = pd.to_numeric(se_df['idx_prix_conso'])
se_df['idx_conf_conso'] = pd.to_numeric(se_df['idx_conf_conso'])
# Filling the `tx_var_emploi` NaN with trimestrial data extrapolation
col_label =  "tx_var_emploi"  # Column label of the datapoint
missing_map = {4 : -0.1, 8 : -0.2,  9 : -0.2, 20 : -3, 21 : -3}
se_df['tx_var_emploi'].fillna(missing_map, inplace=True)
display(se_df)
```

```python
# Perform linear interpolation to the rest of the missing Values
se_df.interpolate(method='linear', inplace=True)
```

```python
#display the daraframe and save it as a new csv file

#display(se_df) 
se_df.to_csv("../data/processed/socio_eco_processed.csv", index=False)
```
